```python
import eval
import pickle
import motmetrics as mm
import numpy as np
from eval import bbox_iou
```

```python
evaluator = eval.BboxTrackEvaluator(0.5)
i = 0
print(f'Evaluating sequence {i}')
dt_result_txt_path = f'temp/{i}/results.txt'
dt_sequence = evaluator.gen_dt_target_sequence(dt_result_txt_path)
gt_sequence_path = f'C:/Users/yliu60/Documents/GitHub/embedding_tracking/dataset/6_shapes/test/seq_{i}/sequence.pickle'
with open(gt_sequence_path, 'rb') as handle:
    gt_sequence = pickle.load(handle)

gt_sequence = evaluator.gen_gt_target_sequence(gt_sequence)
acc = mm.MOTAccumulator(auto_id=True)
for i in range(len(dt_sequence)):
    dt_frame = dt_sequence[i]
    gt_frame = gt_sequence[i+1]
    num_dt = len(dt_frame)
    num_gt = len(gt_frame)
    dt_ids = [x.id for x in dt_frame]
    gt_ids = [x.id for x in gt_frame]
    dist_matrix = np.zeros((num_gt, num_dt))
    for j in range(num_gt):
        for k in range(num_dt):
            gt_target = gt_frame[j]
            dt_target = dt_frame[k]
            iou = bbox_iou(gt_target.bbox, dt_target.bbox)
            if iou < evaluator.iou_threshold:
                dist = np.nan
            else:
                dist = 1 - iou
            dist_matrix[j, k] = dist
    acc.update(gt_ids, dt_ids, dist_matrix)
    break
```
